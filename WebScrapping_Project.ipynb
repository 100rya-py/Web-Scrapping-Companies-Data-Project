# Scrapping Data from a Real Website + Pandas

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### Sending Requests to Website and Parsing HTML

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
}
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.text, "html.parser")
```

```python
print(soup)
```

### Find Table Data

```python
soup.find('table') 
```

```python
soup.find_all('table')[0]
```

```python
table=soup.find('table',class_="wikitable sortable")
table
```

```python
world_titles=soup.find_all('table')[0].find_all('th') 
world_titles
```

```python
world_table_title = [title.text.strip() for title in world_titles]
world_table_title
```

```python
df=pd.DataFrame(columns=world_table_title)
df
```

```python
column_data=soup.find_all('table')[0].find_all('tr')[1:]
column_data
```

### Construting a Table from the Table Data

```python
for row in column_data:
    row_data=row.find_all('td')
    world_table_data=[data.text.strip() for data in row_data]
    length=len(df)
    df.loc[length]=world_table_data
```
